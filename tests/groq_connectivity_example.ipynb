# Groq Connectivity via `src.llm`

This notebook validates Groq connectivity using the project LLM module (`src.llm`) instead of the raw SDK client.

```python
from __future__ import annotations

import sys
import os
from pathlib import Path

repo_root = Path.cwd().resolve()
if not (repo_root / 'src').exists() and (repo_root.parent / 'src').exists():
    repo_root = repo_root.parent
if str(repo_root) not in sys.path:
    sys.path.insert(0, str(repo_root))

from src.llm import LLMFactory
```

```python
def load_env_file(path: Path) -> None:
    if not path.exists():
        print(f"File not found: {path}- {path.exists()}")
        return
    for raw in path.read_text(encoding='utf-8').splitlines():
        line = raw.strip()
        if not line or line.startswith('#') or '=' not in line:
            continue
        key, value = line.split('=', 1)
        os.environ.setdefault(key.strip(), value.strip().strip('"').strip("'"))


repo_root = Path.cwd().resolve()
env_path = Path("../.env")
load_env_file(env_path)

api_key = os.getenv('GROQ_API_KEY', '').strip()
print('GROQ_API_KEY present:', bool(api_key))
print('ENV path:', env_path)
```

```python
if not api_key:
    raise RuntimeError('GROQ_API_KEY is not set in environment or .env file')

llm = LLMFactory.build_groq_llm(
    model_name='llama-3.1-8b-instant',
    api_key=api_key,
    max_new_tokens=256,
    temperature=1.0,
)

response = llm.generate(
    system_prompt='You are a helpful assistant.',
    user_prompt='Who is the chief minister of west bengal',
)

if not response.strip():
    raise RuntimeError('Groq response was empty')

print(response)
```
